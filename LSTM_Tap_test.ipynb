```python
import json
import numpy as np
from matplotlib.pyplot import plot
```

```python
with open ('testTap.txt') as f:
    data = json.load(f)

valid_y=np.empty((0,1))
valid_x = np.empty((0,3))
for i in range(len(data)):
    p=data[i]['p']
    d=data[i]['d']
    for j in range(len(p)):
        y=float(p[j])
        valid_y=np.append(valid_y,np.array(y).reshape([1,1]),axis=0)
        x = float(d[j][0]['x'])
        y = float(d[j][0]['y'])
        z = float(d[j][0]['z'])
        valid_x = np.append(valid_x,np.array([x,y,z]).reshape([1,3]),axis=0)
```
