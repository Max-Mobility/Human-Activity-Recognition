```python
import keras
from keras.layers import Input, Conv1D,Activation,MaxPooling1D,Dense,Flatten,UpSampling1D
from keras.layers.normalization import BatchNormalization
from keras.models import Model,load_model
import tensorflow as tf
import numpy as np
from matplotlib.pyplot import plot
import matplotlib.pyplot as plt
```

```python
gyroModel=load_model('gyroNorModel_64.h5')
linearAccModel=load_model('linearAccNorModel_64.h5')
gravityModel=load_model('gravityNorModel_64.h5')
gameVecModel=load_model('gameVecNorModel_64.h5')
gyroModelFFT=load_model('gyroNorModelFFT_64.h5')
linearAccModelFFT=load_model('linearAccNorModelFFT_64.h5')
gravityModelFFT=load_model('gravityNorModelFFT_64.h5')
gameVecModelFFT=load_model('gameVecNorModelFFT_64.h5')
```

```python
data= np.load('watch_norm_64.npy')
gyro=data[:,:,0:3]
linearAcc=data[:,:,3:6]
gravity=data[:,:,6:9]
gameVec=data[:,:,9:13]

gyro_fft=data[:,:,13:16]
linearAcc_fft=data[:,:,16:19]
gravity_fft=data[:,:,19:22]
gameVec_fft=data[:,:,22:26]
dataLength=64
```

```python
sample = 12
gyroy = gyroModel.predict(gyro[sample].reshape(1,dataLength,3))
fig=plt.figure(figsize=(20,7))
for channel in range(3):
    plt.subplot(1,3,channel+1)
    plot(gyro[sample,:,channel],'b')
    plot(gyroy[0,:,channel],'r')
linearAccy = linearAccModel.predict(linearAcc[sample].reshape(1,dataLength,3))
fig=plt.figure(figsize=(20,7))
for channel in range(3):
    plt.subplot(1,3,channel+1)
    plot(linearAcc[sample,:,channel],'b')
    plot(linearAccy[0,:,channel],'r')
gravityy = gravityModel.predict(gravity[sample].reshape(1,dataLength,3))
fig=plt.figure(figsize=(20,7))
for channel in range(3):
    plt.subplot(1,3,channel+1)
    plot(gravity[sample,:,channel],'b')
    plot(gravityy[0,:,channel],'r')
gameVecy = gameVecModel.predict(gameVec[sample].reshape(1,dataLength,4))
fig=plt.figure(figsize=(20,7))
for channel in range(4):
    plt.subplot(1,4,channel+1)
    plot(gameVec[sample,:,channel],'b')
    plot(gameVecy[0,:,channel],'r')
```

```python
sample = 32
gyro_ffty = gyroModelFFT.predict(gyro_fft[sample].reshape(1,dataLength,3))
fig=plt.figure(figsize=(20,7))
for channel in range(3):
    plt.subplot(1,3,channel+1)
    plot(gyro_fft[sample,:,channel],'b')
    plot(gyro_ffty[0,:,channel],'r')
linearAcc_ffty = linearAccModelFFT.predict(linearAcc_fft[sample].reshape(1,dataLength,3))
fig=plt.figure(figsize=(20,7))
for channel in range(3):
    plt.subplot(1,3,channel+1)
    plot(linearAcc_fft[sample,:,channel],'b')
    plot(linearAcc_ffty[0,:,channel],'r')
gravity_ffty = gravityModelFFT.predict(gravity_fft[sample].reshape(1,dataLength,3))
fig=plt.figure(figsize=(20,7))
for channel in range(3):
    plt.subplot(1,3,channel+1)
    plot(gravity_fft[sample,:,channel],'b')
    plot(gravity_ffty[0,:,channel],'r')
gameVec_ffty = gameVecModelFFT.predict(gameVec_fft[sample].reshape(1,dataLength,4))
fig=plt.figure(figsize=(20,7))
for channel in range(4):
    plt.subplot(1,4,channel+1)
    plot(gameVec_fft[sample,:,channel],'b')
    plot(gameVec_ffty[0,:,channel],'r')
```
